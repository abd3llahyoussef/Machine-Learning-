```python
from keras import layers
from keras.preprocessing import sequence
from keras.datasets import imdb
import tensorflow as tf
import numpy as np
```

```python
Vocap_Size =88584
Max_Len = 250

(x_train,y_train),(x_test,y_test) = imdb.load_data(num_words=Vocap_Size)
```

```python
x_train = sequence.pad_sequences(x_train,Max_Len)
x_test = sequence.pad_sequences(x_test,Max_Len)
```

```python
x_train[0]
```

```python
# RNN Model
model = tf.keras.Sequential([
    layers.Embedding(Vocap_Size,64),
    layers.LSTM(64),
    layers.Dense(1,activation="sigmoid")
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.0001),loss="binary_crossentropy",metrics=["accuracy"])
model.fit(x_train,y_train,epochs=8,batch_size=64,validation_split=0.2)
```

```python
loss , acc = model.evaluate(x_test,y_test)
```

```python
#Encode A Word
wordIndex = imdb.get_word_index()

def convert (text):
  seq = tf.keras.preprocessing.text.text_to_word_sequence(text)
  seq = [wordIndex[word] if word in wordIndex else 0 for word in seq]
  return sequence.pad_sequences([seq],Max_Len)[0]

seq = "I Love This Move"
seq = convert(seq)
```

```python
# Reverse A Word
reverse = {value:key for (key,value) in wordIndex.items()}

def convertBack(integers):
  Pad = 0
  text = ""
  for num in integers:
    if num != Pad:
      text += reverse[num] + " "
  return text[:-1]

print(convertBack(seq))
```

```python
# prediction
def predict(text):
  seq = convert(text)
  pred = np.zeros((1,250))
  #print(pred)
  pred[0] = seq
  #print(pred[0])
  #print(pred)
  res = model.predict(pred)

  return res[0]

review = "That movie was! really loved it and would great watch it again because it was amazingly great"
print(predict(review))

review = "that movie really sucked. I hated it and wouldn't watch it again. Was one of the worst things I've ever watched"
print(predict(review))
```
